```python
import pandas as pd
import os
from datetime import datetime
```

```python
xls_list = os.listdir("D:\推广\雪肌精\日报\流量访客分析")
```

```python
df_first = pd.read_excel(r"D:/推广/雪肌精/日报/流量访客分析/"+xls_list[0],header=4)
```

```python
df_traffic = df_first.head(df_first[df_first.isnull().T.any()].index.to_list()[0])
df_traffic["日期"] = datetime.strptime(xls_list[0].replace(".xls",""),"%Y-%m-%d")
df_buy = df_first.tail(len(df_first)-df_first[df_first.isnull().T.any()].index.to_list()[0]-1)
df_buy["日期"] = datetime.strptime(xls_list[0].replace(".xls",""),"%Y-%m-%d")
```

```python
for file in xls_list[1:]:
    df = pd.read_excel("D:/推广/雪肌精/日报/流量访客分析/"+file,header=4)
    df_temp_traffic = df.head(df[df.isnull().T.any()].index.to_list()[0])
    df_temp_traffic["日期"] = datetime.strptime(file.replace(".xls",""),"%Y-%m-%d")
    df_temp_buy = df.tail(len(df)-df[df.isnull().T.any()].index.to_list()[0]-1)
    df_temp_buy["日期"] = datetime.strptime(file.replace(".xls",""),"%Y-%m-%d")
    df_traffic = pd.concat([df_traffic,df_temp_traffic])
    df_buy = pd.concat([df_buy,df_temp_buy])
```

```python
df_traffic = df_traffic[["日期","统计终端","地域","访客数","下单转化率"]]
df_buy = df_buy[["日期","统计终端","地域","访客数","下单转化率"]]
df_traffic.to_excel("D:/推广/雪肌精/日报/访客来源.xlsx")
df_buy.to_excel("D:/推广/雪肌精/日报/购买来源.xlsx")
```
