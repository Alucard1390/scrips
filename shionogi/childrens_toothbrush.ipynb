```python
import pandas as pd
import numpy as np
```

```python
sales_sheet = pd.read_excel(r"D:\儿童牙刷.xlsx",sheet_name="热卖商品")
brands_sheet = pd.read_excel(r"D:\儿童牙刷.xlsx",sheet_name="热卖品牌")
```

```python
brands = brands_sheet["品牌名称"].to_list()
```

```python
print(brands)
```

```python
brands_new = []
for i in brands:
    if "（" in str(i):
        k = str(i)[:str(i).find("（")]
    else:
        k = str(i)
    brands_new.append(k)
    brands_new.append(k.upper())
    brands_new.append(k.lower())
```

```python
brands_new
```

```python
def pairing(shop_name,brands_list):
    for j in brands_list:
        if j in shop_name.upper() or j in shop_name.lower():
            return j
        else:
            continue
```

```python
sales_sheet["品牌名称"] = sales_sheet["店铺名称"].apply(lambda x:pairing(x,brands_list=brands_new))
```
